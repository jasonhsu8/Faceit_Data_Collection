```python
import requests
import pandas as pd
import numpy as np
import json
import os
from os import listdir
from os.path import isfile, join
```

```python

onlyfiles = [f for f in listdir("./MyFaceitData/Match Stats") if isfile(join("./MyFaceitData/Match Stats", f))]
```

```python
def rename_cols(text: str):
    text = text.split(".")[1]
    text = text.split(" ")
    text = "".join(list(map(str.capitalize, text)))
    return text
```

```python
def create_df_from_json():
    df = pd.DataFrame()
    for ID in onlyfiles:
        path = f"./MyFaceitData/Match Stats/{ID}"
        with open(path, "r") as file:
           obj = json.load(file)["items"]
           obj = pd.json_normalize(obj)
        obj = obj.rename(columns=rename_cols)
        df = pd.concat([df, obj], ignore_index=True)
    return df
```

```python
df = create_df_from_json()
```

```python
df.to_pickle("JJStats.pkl", compression="bz2")
```
